```python
pattern = 'hostname:%s'%host
facets = 'os,org,port'
search_url = 'https://api.shodan.io/shodan/host/search?key=%s&query=%s&facets=%s'%(api_key,pattern,facets)

res = requests.get(search_url)

print(res.json()['total'])
r = res.json()['matches']

for elem in r:
    print("%s : %s"%(elem['hostnames'],elem['ip_str']))
    for key in elem.items():
        print("%s : %s"% (elem['hostnames'], key))
        
sorted(set(match['asn'] for match in r if 'asn' in match), key=lambda x:int(x.replace('AS','')),reverse=True)
```
